```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
chrome_driver_path = '/Users/jackgoettle/Downloads/chromedriver'
browser = webdriver.Chrome(executable_path = chrome_driver_path)
base_url = 'https://stats.nba.com/players/boxscores-usage/?Season='
end_url = '&SeasonType=Regular%20Season'
```

```python
# convert attrbs to int
def to_int(num) :
    try :
        return int(num)
    except :
        return num

# convert attrbs to float
def to_float(num) :
    try :
        return float(num)
    except :
        return num
```

```python
def check_page(browser) :
    table = browser.find_element_by_class_name('nba-stat-table__overflow')
    if table == None :
        return False
    return table

def get_next_page(browser) :
    browser.implicitly_wait(10)
    browser.find_element_by_class_name('stats-table-pagination__next').click()
    return check_page(browser)

def scrape(season) :
    url = base_url + season + end_url
    browser.get(url)
    browser.implicitly_wait(100)
    players = []
    table = check_page(browser)
    count = 0
    prev = ''
    while table is not None : 
        name = ''
        lines = table.text.split('\n')
        if prev == lines[1] :
            break
        prev = lines[1]
        for idx, stats in enumerate(lines):
            if idx == 0 :
                continue
            elif idx % 2 == 1 :
                name = stats
            else :
                arr = stats.split(' ')
                matchup = arr[1] + arr[2] + arr[3]
                player = {'PLAYER':name, 'TEAM':arr[0], 'MATCH UP':matchup, 'GAME DATE':arr[4],
                          'W/L':arr[5], 'MIN':to_int(arr[6]), 'USG%':to_float(arr[7]), 
                          '%FGM':to_float(arr[8]), '%FGA':to_float(arr[9]), '%3PM':to_float(arr[10]),
                          '%3PA':to_float(arr[11]), '%FTM':to_float(arr[12]), '%FTA':to_float(arr[13]),
                          '%OREB':to_float(arr[14]), '%DREB':to_float(arr[15]), '%REB':to_float(arr[16]),
                          '%AST':to_float(arr[17]), '%TOV':to_float(arr[18]), '%STL':to_float(arr[19]),
                          '%BLK':to_float(arr[20]), '%BLKA':to_float(arr[21]), '%PF':to_float(arr[22]),
                          '%PFD':to_float(arr[23]), '%PTS':to_float(arr[24])
                          }
                players.append(player)
        count += 1
        print(count)
        table = get_next_page(browser)
    return pd.DataFrame(players)
```

```python
yr = None
with open('/Users/jackgoettle/DFS/players_traditional_2019-20.csv', 'r') as f :
    yr = pd.read_csv(f)
```

```python
yrs = ['2014-15', '2015-16', '2016-17', '2017-18', '2018-19']
for yr in yrs : 
    df = scrape(yr)
    csv = 'players_usage_' + str(yr) + '.csv'
    df.to_csv(csv, index=False)
```
