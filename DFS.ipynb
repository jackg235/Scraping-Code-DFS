```python
import random 
import numpy as np
import pandas as pd
import datetime
import requests
from bs4 import BeautifulSoup
from datetime import timedelta
import ssl
```

```python
years = [i + 2014 for i in range(7)]
months = [i + 1 for i in range(12) if i < 6 or i > 8]
days = [i + 1 for i in range(31)]
base_url = 'http://rotoguru1.com/cgi-bin/hyday.pl?game=fd&'
```

```python
def scrape(mon, day, year) : 
    url = base_url + 'mon=' + str(mon) + '&day=' + str(day) + '&year=' + str(year)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    elements = soup.find_all('td')
    found = False
    count = 0
    name = ''
    salary = 0
    points = 0
    
    players = []
    for el in elements :
        el = el.text
        if el == 'PG' or el =='SG':
            found = True
        if found :
            if count % 9 == 1 :
                name = el             
                if name.endswith('^') :
                    name = name[:len(name) - 1]
            elif count % 9 == 2 :
                points = el
            elif count % 9 == 3 :
                salary = el[1:].replace(',', '')
                if 'Unlisted' in name :
                    break
                if 'Forwards' not in name and 'Centers' not in name :
                    player = {'name' : name, 'points' : float(points), 'salary' : int(salary)}
                    players.append(player)
            count+=1
    df = pd.DataFrame(players)
    if (len(df) == 0) :
        return None
    df['date'] = datetime.datetime(year, mon, day)
    return df
```

```python
master = pd.DataFrame(columns = ['name', 'points', 'salary', 'date'])
for y in years :
    for m in months : 
        for d in days :
            print('year= ' + str(y) + 'mon=' + str(m) + 'day=' + str(d))
            df = scrape(m, d, y)
            if df is not None : 
                master = master.append(df, ignore_index = True)
```

```python
master.to_csv('dfs.csv', index = False)
```
